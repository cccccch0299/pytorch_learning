nn.Module 做了三件事：

* 管理参数
* 定义 forward
* 支持自动求导

手动实现2x + b 

```python
import torch
import torch.nn as nn

class myModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.w = nn.Parameter(torch.tensor(2.0))
        self.b = nn.Parameter(torch.tensor(1.0))
    
    def forward(self,x):
        return self.w * x + self.b
    
```

```python
model = myModel()
x = torch.tensor([[1.0],[2.0],[3.0],[4.0]])
y = model(x)
print(y)
```

## 实战
拟合函数 y = 2x+3

```python
import torch
import torch.nn as nn

class mymodel1(nn.Module):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        # # bias=True 是默认的，它会自动创建一个 b 参数
        # self.linear = nn.Linear(100, 100) 
        
        
        self.w = nn.Parameter(torch.tensor(0.0))
        self.b = nn.Parameter(torch.tensor(0.0))

    def forward(self,x):
        #return self.linear(x)
        return self.w * x + self.b
```

```python
x = torch.randn(64,100)
y_true = 2*x+3
mymodel = mymodel1()
lossfn = nn.MSELoss()
optimizer = torch.optim.SGD(mymodel.parameters(),lr = 0.01)
for epoch in range(1000):
    y_pred = mymodel(x)
    loss = lossfn(y_pred,y_true)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    if(epoch % 100 == 0):
        print(loss.item())

print(mymodel.w.item())
print(mymodel.b.item())
# print(mymodel.linear.weight.data)
# print(mymodel.linear.bias.data)
```
