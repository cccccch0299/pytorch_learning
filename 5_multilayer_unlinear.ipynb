## 多层神经网络 和 激活函数

### 激活函数
* relu              f(x) = max(0,x)
* sigmoid           值域(0,1)
* tanh              值域(-1,1)

### 多层神经网络


```python
import torch.nn as nn

class myMLP(nn.Moudle):
    def __init__(self):
        #先要继承
        super.init()
        self.fcn1 = nn.Linear(10,1)
        self.relu = nn.ReLU()
        self.fcn2 = nn.Linear(1,10)
    
    def forward(self,x):
        x = self.fcn1(x)
        x = self.relu(x)
        x = self.fcn2(x)
        return x
```

**简洁写法**

```python
class myMLP2(nn.Moudle):
    def __init__(self):
        #先要继承
        super.init()
        self.model = nn.Sequential(
            nn.Linear(1,10),
            nn.ReLU(),
            nn.Linear(10,1)
        )
    
    def forward(self,x):
        y = self.model(x)
        return y
```
