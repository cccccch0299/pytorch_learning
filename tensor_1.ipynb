## 一、创建tensor
torch.arang 用于生成一个指定范围内的张量，其元素是按一定步长排列的。

```python
import torch
a = torch.tensor([2,3,4])
a1 = torch.arange(6)
a2 = torch.arange(start = 0,end = 10,step=2)#左开右闭
print(a)
print(a1)
print(a2)
```

## 三、基本运算

### 矩阵乘法！！！
@ 和 matmul

**最后两维相乘** 形状 (m, n, k) @ (m, k, p) -> 结果 (m, n, p)

```python
i = torch.rand(3,4)
j = torch.rand(4,8)
print(i@j)
torch.matmul(i,j)
```

## 四、reshape/view(变形)
**reshape：更安全**

**view：更快（但要求内存连续）**

转置 .t() 操作返回的是一个“非连续”的张量视图

```python
k = torch.arange(6)
l = k.reshape(3,2)
m = k.view(2,3)
print(l)
print(m)

x = torch.randn(2,3)
y = x.t()   # 转置（内存不连续）
y.view(6)   # ❌ 会报错
#正确写法
y = y.contiguous().view(6)
```

## 五、squeeze / unsqueeze(非常常用！！！！)
squeeze 去掉维度 (只能移除大小为 1 的维度)

unsqueeze 增加维度

```python
n = torch.randn(1,3,2)
n1 = torch.randn(3,2)
n2 = torch.randn(2,3,2) #无法去除
o = n.squeeze()
print(o)
p = n1.unsqueeze(1) #参数是增加的维度位置 0 就是变成(1,3,2) 1就是变成(3,1,2)以此类推
print(p)
p.shape
```

### 高级索引

```python
# 布尔索引
scores = torch.tensor([[0.1, 0.9, 0.2],
                       [0.8, 0.3, 0.4]])

# 找出所有大于 0.5 的元素的位置
mask = scores > 0.5
# mask 是布尔矩阵，直接索引会返回所有满足条件的元素（展平为一维）
high_scores = scores[mask] 
# 结果: tensor([0.9, 0.8])
print(high_scores)

# 花式
x = torch.tensor([[1,2],
                  [3,4]])

idx = torch.tensor([0,1])

print(x[idx, idx])  # [1,4]
```

## NumPy ↔ Torch 转换（重要‼️）

numpy --> tensor

torch.from_numpy()  零拷贝 (Zero-copy)：转换速度极快，不复制数据。

torch.tensor()   数据拷贝 (Copy)：会分配新内存并复制数据，转换速度稍慢（对于大数组）。

tensor --> numpy

xxx.numpy()


```python
import numpy as np
r = np.array([1,2,3])
s1 = torch.from_numpy(r) #共享内存
s2 = torch.tensor(r)
print(s1)
print(s2)
s3 = s2.numpy()
s1[1] = 4
print(r)
```
