## 自定义dataset

```python
from torch.utils.data import Dataset
import torch

class myDataset(Dataset):
    def __init__(self):
        self.x = torch.randn(100,1)
        self.y = 2*self.x +3
    
    def __len__(self):
        return len(self.x)
    
    def __getitem__(self,index):
        return self.x[index], self.y[index]
        
```

## DataLoader
shuffle是否打乱
batch_size一次性送入多少数据

```python
from torch.utils.data import DataLoader
dataset = myDataset()
loader = DataLoader(dataset,shuffle=True,batch_size=16)
for x,y in loader:
    print(x.shape)
```
